```python
# checking thte null values in percentage form
def missing_data(dataset):
    total=dataset.isnull().sum().sort_values(ascending=False)
    percentage=((dataset.isnull().sum()/len(dataset.index)*100).sort_values(ascending=False))
    missing_data=pd.concat([total,percentage],axis=1,keys=["total","percentage"])
    return(missing_data.head(30))
print(missing_data(dataset))
```

```python
# dropping the insignificant variables which are not required for model building
dataset=dataset.drop(["id","member_id","earliest_cr_line","title","zip_code","policy_code"],axis=1)
```

```python
# converting the object variables into integer
colname=[]
for i in dataset.columns:
    if dataset[i].dtype=="object" or dataset[i].dtype=="bool":
        colname.append(i)
print(colname)
```

```python
dataset_new=dataset.drop(["issue_d"],axis=1)
```

```python
# converting the object variables into integer
colname=[]
for i in dataset_new.columns:
    if dataset_new[i].dtype=="object" or dataset[i].dtype=="bool":
        colname.append(i)
print(colname)
```

```python
# labelling the character varibale into numerical variables
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for a in colname:
    dataset_new[a]=le.fit_transform(dataset_new[a])
```

```python
dataset_new.head()
```

```python
dataset_new["issue_d"]=dataset["issue_d"]
```

```python
dataset_new.head()
```

```python
# converting the issue_d variable into datetime format
dataset_new["issue_d"]=pd.to_datetime(dataset_new["issue_d"])
dataset_new.head()
```

```python
# creating the train & test data
train=dataset_new[dataset_new["issue_d"]<"2015-06-01"]
test=dataset_new[dataset_new["issue_d"]>="2015-06-01"]
train.head()
```

```python
X_train=train.drop(["default_ind","issue_d"],axis=1)
Y_train=train["default_ind"]
X_test=test.drop(["default_ind","issue_d"],axis=1)
Y_test=test["default_ind"]
```

```python
# building the logistic model
rom sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,Y_train)
y_pred=classifier.predict(X_test)
```

```python
# evaluation of model
print(classifier.intercept_)
print(classifier.coef_)

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

print(confusion_matrix(Y_test,y_pred))

print(accuracy_score(Y_test,y_pred))

print(classification_report(Y_test,y_pred)) 
```

```python
from sklearn import metrics

fpr,tpr,thresholds=metrics.roc_curve(Y_test,y_pred)
auc_score=metrics.auc(fpr,tpr)
```

```python
auc_score
```

```python
import matplotlib.pyplot as plt
plt.title("Receiver Operating Characteristic")
plt.plot(fpr,tpr,"g",label=auc_score) #"g" indicates green colour
plt.plot([0,1],[0,1],"b")
plt.legend(loc="lower right")# this code is for location of label
plt.xlabel("false postive rate")
plt.ylabel("true positive rate")
plt.show()
```

```python
# checking the probability values of each observation
Y_pred_prob=classifier.predict_proba(X_test)
print(Y_pred_prob)
```

```python
import numpy as np
for a in np.arange(0.1,0.51,0.01):
    predict_mine=np.where(Y_pred_prob[:,1]>a,1,0)
    cfm=confusion_matrix(Y_test,predict_mine)
    total_err=cfm[0,1]+cfm[1,0]
    print("Errors at threshold",a,":",total_err,", type 2 error :",
         cfm[1,0],", type 1 error",cfm[0,1])
```

```python
Y_pred_class=[]
for x in Y_pred_prob[:,1]:
    if x > 0.49:
        Y_pred_class.append(1)
    else:
        Y_pred_class.append(0)
print(Y_pred_class)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cfm=confusion_matrix(Y_test,Y_pred_class)
print(cfm)
acc=accuracy_score(Y_test, Y_pred_class)
print("Accuracy of the model: ",acc)
print(classification_report(Y_test, Y_pred_class))
```
